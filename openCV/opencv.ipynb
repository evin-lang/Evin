```
import cv2
import os

def save_all_frames(video_path, dir_path, basename, ext='jpg'):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        return

    os.makedirs(dir_path, exist_ok=True)
    base_path = os.path.join(dir_path, basename)

    digit = len(str(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))))

    n = 0

    while True:
        cap.set(cv2.CAP_PROP_POS_FRAMES,n*100)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite('{}_{}.{}'.format(base_path, str(n).zfill(digit), ext), frame)
            n += 1
        else:
            return

save_all_frames('/content/drive/My Drive/openCV/movie/input/よびのり.mp4', '/content/drive/My Drive/openCV/movie/output', 'video_img', 'png')
```

```
import os, sys
import numpy as np
import cv2
import matplotlib
from matplotlib import pylab as plt
from mpl_toolkits.mplot3d import Axes3D

class ImageColorProcessing(object):

  def __init__(self, image_path=None):
        self.image_path = image_path
        self.image = self.read_image(self.image_path)
  def read_image(self, image_path):
      image = cv2.imread(image_path)
      if image is None:
          print("Image[{}] can not open.".format(image_path))
          sys.exit(1)
      return image
      
  def bgr_to_rgb_to_flat(self):
      rgb_image = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB) # BGR -> RGB
      height, width, channel = rgb_image.shape
      rgb_flat = rgb_image.reshape((height*width, channel)) # 平板化
      return rgb_flat

  def duplication_rgb(self):
      flat = self.bgr_to_rgb_to_flat()
      duplication, counts = np.unique(flat, return_counts=True, axis=0)
      return duplication, counts.reshape((counts.shape[0], 1))

  def plot3d(self, x, y, z, size, color):
      fig = plt.figure()
      ax = fig.add_subplot(111, projection="3d")
      ax.scatter(x, y, z, s=size, c=color)
      ax.set_xlim([0, 255])
      ax.set_ylim([0, 255])
      ax.set_zlim([0, 255])
      ax.set_xlabel("R")
      ax.set_ylabel("G")
      ax.set_zlabel("B")
      plt.show()

  def rgb_plot3d(self):
      duplication, counts = self.duplication_rgb()
      red, green, blue = np.hsplit(duplication, [1, 2])
      self.plot3d(red, green, blue, counts, duplication/255.)
  
  def rgbData_to_csv(self):
      duplication, counts = self.duplication_rgb()
      rgbData = np.hstack([duplication, counts])
      np.savetxt("rgb.csv", rgbData, delimiter=",", header="r,g,b,counts")

def main():
  files =[file for file in find_all_files(url+'/output'+testcase)]
  images = [cv2.imread(file) for file in sorted(files,key=lambda file:int(re.search('\d+',re.sub(url+'/input'+testcase+'/video_img_','',file)).group()))]
  ICP = ImageColorProcessing(image_path=url+'/output'+testcase+'/diff3-4.png') # 「***」は処理したい画像のパスを
  print(files[0])
  ICP.rgb_plot3d() # 視覚化
    #ICP.rgbData_to_csv() # データ保存

if __name__ == "__main__": main()
```

```
from collections import Counter

def calculate_mode(data):
    c = Counter(data)
    # すべての要素とその出現回数を取り出します。
    freq_scores = c.most_common()
    #c.most_common内の最も多い要素[0]の最大出現回数[1]を[0][1]で指定
    max_count = freq_scores[0][1]

    modes = []
    #出現回数と最大出現回数が等しいかを確認します。
    for num in freq_scores:
        if num[1] == max_count:
            modes.append(num[0])
    return(modes)

if __name__ == '__main__':
    data = [100,200,300,400,500,500,600,700,800,800,10,10]
    modes = calculate_mode(data)

    print('もっとも頻度の高い数は:')
    for mode in modes:
        print(mode)
```

最頻値法
長所
*   思ったより安定し、きれいな画像を生成できる
*   差分計算の時間がいらないため早い


短所
*   時間がかかりすぎ(約７０秒)
> 実はもっと増えそうな予感がする…
*   複数の最頻値が存在した場合、どの最頻値をとればいいのかがわからない


観察
*   初めのほうの最頻値を取ろうが、あとのほうの最頻値を取ろうが関係がない。
> この対策として一定の個数を得られたらそれを画素の値として認定するというもの
*   最頻値を集めた配列の中の数字の範囲が１０以上のものは約８０００個(総数の約１％)
*   Overの数の変化を見極めたほうがおもしろい？？
> よく動く場所、変化の激しい場所として記録できる
*   240など閾値を定めて無理矢理行うという場合
> B，G，Rという風にわけて行うと特定の色が強く出る場合がある
>> 3つの画素を評価の対象としBGRで分割しない。


改善案
*    ある程度、連続した値が来たら塗り替える法
> 3回ぐらい±１０くらいの範囲で連続したらその値をとるというもの
> 流れ
>>１．値を取得、cntを１する
>>
>>２．±１０の範囲にいるかを検索。ないなら値を更新するために１へ
>>
>>３．cntを＋１する。２へ　cnt==3ならば 4へ
>>
>>４．その値が適切であると判断し、その値に塗り替える
>>
>>何度も塗り替えるのを防ぐために0に初期化せず±１０を超えるものがくるまでcntをつづける
> 値が違う時点で放棄できる
*　　周りを一括に計算し比較する法
> ある程度、連続した値が来たら塗り替える法を改善したもの
> 流れ
> 1. ６＊６ぐらいの正方形を画像から取得する。

```
start = time.time()
files =[file for file in find_all_files(url+'/input'+testcase)]
images = [cv2.imread(file) for file in sorted(files,key=lambda file:int(re.search('\d+',re.sub(url+'/input'+testcase+'/video_img_','',file)).group()))]
width, height, color = images[0].shape
over = 0
over240 = 0
for i in range(0,width):
  for j in range(0,height):
    pixels = [pix.tolist() for pix in [img[i,j,:] for img in images]]
    f = 0
    for cel in pixcels:
      for pix in cel:
        if pix > 250:
          f += 1
          val = cel
          
    for k in range(0,color):
      data = [pix[k] for pix in pixels]
      c = Counter(data)
      freq = c.most_common()
      #print(freq)
      #print("freq")
      #print(freq)
      #print("freq0")
      max_count = freq[0][1]
      modes = []
      for num in freq:
        if num[1] == max_count:
          modes.append(num[0])
      if abs(max(modes)-min(modes))>30:
        over+=1
        #print('over10:{}'.format(abs(max(modes)-min(modes))))

      if f > 3:
        images[0][i,j] = val
        over240 +=1
        #print('over3')
      else:
        images[0][i,j][k] = max(modes)
      #print(modes)
cv2.imwrite(url+'/output'+result+'/freqcombine_over3_250.png',images[0])
print("finish:{:.2f}[sec]".format(time.time()-start))
print("number of over10:{}".format(over))
print("number of over240:{}".format(over240))
print("over rate:{:.2f}".format((over/(width*height))*100))
```
